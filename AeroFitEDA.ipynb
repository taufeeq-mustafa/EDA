```python
numerical_columns = ['Age', 'Education', 'Usage', 'Fitness', 'Income', 'Miles']

for column in numerical_columns:
    Q1 = df[column].quantile(0.25)  # First quartile (25th percentile)
    Q3 = df[column].quantile(0.75)  # Third quartile (75th percentile)
    IQR = Q3 - Q1                  # Interquartile Range
    lower_bound = Q1 - 1.5 * IQR    # Lower bound
    upper_bound = Q3 + 1.5 * IQR    # Upper bound

    # Identifying outliers
    outliers = df[(df[column] < lower_bound) | (df[column] > upper_bound)]

    print(f"Column: {column}")
    print(f"  Lower Bound: {lower_bound:.2f}, Upper Bound: {upper_bound:.2f}")
    print(f"  Number of Outliers: {len(outliers)}")
    print("-" * 40)
```

```python
product_percentage = (df['Product'].value_counts(normalize=True) * 100).round(2).astype(str) + '%'
print(product_percentage)
```

```python
male_percentage = (df['Gender'].value_counts(normalize=True)['Male'] * 100).round(2)
print(f"Percentage of Male customers purchasing a treadmill: {male_percentage}%")
```

```python
female_kp781_percentage = ((df['Gender'] == 'Female') & (df['Product'] == 'KP781')).sum() / len(df) * 100
female_kp781_percentage = round(female_kp781_percentage, 2)
print(f"Percentage of Female customers purchasing KP781 treadmill: {female_kp781_percentage}%")
```

```python
kp281_female_prob = (df[(df['Product'] == 'KP281') & (df['Gender'] == 'Female')].shape[0]) / (df['Product'] == 'KP281').sum()
kp281_female_prob = round(kp281_female_prob * 100, 2)
print(f"Probability of a customer being Female given that Product is KP281: {kp281_female_prob}%")
```

```python
age_20_30_percentage = ((df['Age'] >= 20) & (df['Age'] <= 30)).sum() / len(df) * 100
age_20_30_percentage = round(age_20_30_percentage, 2)
print(f"Percentage of customers aged between 20s and 30s: {age_20_30_percentage}%")
```

```python
# Assume low-income as customers earning less than 40,000.
low_income_percentage = (df['Income'] < 40000).sum() / len(df) * 100
low_income_percentage = round(low_income_percentage, 2)
print(f"Percentage of low-income customers purchasing a treadmill: {low_income_percentage}%")
```

```python
high_income_kp781_percentage = ((df['Income'] > 80000) & (df['Product'] == 'KP781')).sum() / len(df) * 100
high_income_kp781_percentage = round(high_income_kp781_percentage, 2)
print(f"Percentage of high-income customers purchasing KP781: {high_income_kp781_percentage}%")
```

```python
high_income_given_kp781 = df[(df['Income'] > 80000) & (df['Product'] == 'KP781')].shape[0] / (df['Product'] == 'KP781').sum() * 100
high_income_given_kp781 = round(high_income_given_kp781, 2)
print(f"Percentage of high-income customers buying treadmill given Product is KP781: {high_income_given_kp781}%")
```

```python
fitness_5_percentage = (df['Fitness'] == 5).sum() / len(df) * 100
fitness_5_percentage = round(fitness_5_percentage, 2)
print(f"Percentage of customers with fitness level 5: {fitness_5_percentage}%")
```

```python
fitness_5_kp781_percentage = ((df['Fitness'] == 5) & (df['Product'] == 'KP781')).sum() / len(df) * 100
fitness_5_kp781_percentage = round(fitness_5_kp781_percentage, 2)
print(f"Percentage of customers with fitness level 5 purchasing KP781: {fitness_5_kp781_percentage}%")
```

```python
partnered_percentage = (df['MaritalStatus'] == 'Partnered').sum() / len(df) * 100
partnered_percentage = round(partnered_percentage, 2)
print(f"Percentage of customers who are partnered using treadmills: {partnered_percentage}%")
```
